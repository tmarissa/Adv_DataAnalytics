```python
import requests
import json
import pandas as pd
```

```python
with open('brewDB_key.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
url = r"https://sandbox-api.brewerydb.com/v2/breweries/?key="
```

```python
response = requests.get(url+api_key)
```

```python
brewdict = {'ID':[], 'Name':[], 'Year of Establishment':[],'Status':[], 'Is it in business?':[],
           'Is it mass owned?':[], 'Is it organic?':[], 'Website':[], 'Description':[]}

for brew in brewdata['data']:
    try:
        brewdict['ID'].append(brew['id'])
    except:
        brewdict['ID'].append(None)
    try:
        brewdict['Name'].append(brew['name'])
    except:
        brewdict['Name'].append(None)
    try:
        brewdict['Year of Establishment'].append(brew['established'])
    except:
        brewdict['Year of Establishment'].append(None)
    try:
        brewdict['Status'].append(brew['status'])
    except:
        brewdict['Status'].append(None)
    try:
        brewdict['Is it in business?'].append(brew['isInBusiness'])
    except:
        brewdict['Is it in business?'].append(None)
    try:
        brewdict['Is it mass owned?'].append(brew['isMassOwned'])
    except:
        brewdict['Is it mass owned?'].append(None)
    try:
        brewdict['Is it organic?'].append(brew['isOrganic'])
    except:
        brewdict['Is it organic?'].append(None)
    try:
        brewdict['Website'].append(brew['website'])
    except:
        brewdict['Website'].append(None)
    try:
        brewdict['Description'].append(brew['description'])
    except:
        brewdict['Description'].append(None)
```

```python
df = pd.DataFrame(brewdict)

df.head()
```
