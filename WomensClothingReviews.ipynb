```python
import pandas as pd

import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from string import punctuation
```

```python
woclodf=pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
woclodf.head()
```

```python
woclodf.shape
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
eng_stopwords = stopwords.words('english')
```

```python
def reviewSentiment(review):
    
    try:
    
        review = review.lower()

        tknz_review = word_tokenize(review)
    
        for token in tknz_review:
            if token in punctuation:
                tknz_review.remove(token)
    
        clean_tokens = []
    
        for token in tknz_review:
            #if token not in eng_stopwords:
            clean_tokens.append(token)
            
        clean_review = ' '.join(clean_tokens)
    
        sid_rev = sid.polarity_scores(clean_review)
    
        r_comp = sid_rev['compound']
    
        return r_comp
    except:
        return None
```

```python
woclodf['Review Sentiment'] = woclodf['Review Text'].apply(reviewSentiment)
```
